```python

###  Visualisation des Données avec Matplotlib et Seaborn

## 1. Introduction à Matplotlib


import matplotlib.pyplot as plt

# Créer un graphique linéaire
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graphique Linéaire')
plt.show()
```

```python
# Ajout de légende et personnalisation
plt.plot(x, y, label='Ligne 1', color='r')
plt.plot(x, [i*2 for i in y], label='Ligne 2', color='g')
plt.legend()
plt.title('Graphique avec plusieurs lignes')
plt.show()
```

```python
## 2. Introduction à Seaborn

import seaborn as sns

# Charger un ensemble de données exemple
tips = sns.load_dataset('tips')

# Créer un diagramme de dispersion
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.title('Pourboires en fonction du montant total de la facture')
plt.show()
```

```python
# Diagramme de distribution
sns.histplot(tips['total_bill'], kde=True)
plt.title('Distribution des Montants Totaux des Factures')
plt.show()
```
